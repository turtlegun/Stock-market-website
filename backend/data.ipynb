```python
import yfinance as yf
import plotly.graph_objects as go

# Define the stock symbols for Sensex and NIFTY 50
sensex_symbol = '^BSESN'
nifty50_symbol = '^NSEI'

# Define the date range
start_date = '2022-01-01'
end_date = '2023-11-01'

# Fetch data for Sensex and NIFTY 50
sensex_data = yf.download(sensex_symbol, start=start_date, end=end_date)
nifty50_data = yf.download(nifty50_symbol, start=start_date, end=end_date)

# Create an interactive line chart
fig = go.Figure()

# Add Sensex data
fig.add_trace(go.Scatter(x=sensex_data.index, y=sensex_data['Close'], mode='lines', name='Sensex'))

# Add NIFTY 50 data


# Customize the layout
fig.update_layout(
    title='Sensex and NIFTY 50 Stock Prices',
    xaxis_title='Date',
    yaxis_title='Price',
    hovermode='x',
)

# Show the interactive plot
fig.show()
```
